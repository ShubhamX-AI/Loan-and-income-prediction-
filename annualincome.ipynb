```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head(5)
```

```python
df2 = pd.concat([df.drop('occupation', axis =1 ), pd.get_dummies(df['education'], prefix='education')], axis=1)
```

```python
df2
```

```python
df2 = pd.concat([df2.drop('marital_status', axis =1 ), pd.get_dummies(df2['marital_status'], prefix='marital_status')], axis=1)
df2 = pd.concat([df2.drop('relationship', axis =1 ), pd.get_dummies(df2['relationship'], prefix='relationship')], axis=1)
df2 = pd.concat([df2.drop('race', axis =1 ), pd.get_dummies(df2['race'], prefix='race')], axis=1)
df2 = pd.concat([df2.drop('workclass', axis =1 ), pd.get_dummies(df2['workclass'], prefix='workclass')], axis=1)
df2 = pd.concat([df2.drop('native_country', axis =1 ), pd.get_dummies(df2['native_country'], prefix='native_country')], axis=1)
df2 = df2.drop('education', axis =1)
```

```python
df2.head(5)
```

```python
df2['gender'] = df2['gender'].apply(lambda x: 1 if x == 'Male' else 0)
df2['income'] = df2['income'].apply(lambda x: 1 if x == '>50K' else 0)
```

```python
df2
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

train, test = train_test_split(df2, test_size=0.2, random_state=42)
```

```python
train
```

```python
train_x = train.drop('income', axis =1)
train_y = train['income']
test_x = test.drop('income', axis =1)
test_y = test['income']
```

```python
forest = RandomForestClassifier(n_estimators=100)
forest.fit(train_x, train_y)
```

```python
forest.score(test_x, test_y)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, forest.predict(test_x))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'n_estimators': [50, 100, 150,],
    'max_depth': [10, 20 , 30 , None],
    'max_features': ['auto', 'sqrt', 'log2'],
    'min_samples_split': [2, 5, 10],
}
grid_search = GridSearchCV(forest, param_grid, cv=5, scoring='accuracy')
grid_search.fit(train_x, train_y)
```

```python
grid_search.best_estimator_
```

```python
forest_best = grid_search.best_estimator_
```

```python
forest_best.score(test_x, test_y)
```

```python
import pickle
pickle.dump(forest_best, open('annual_incomemodel.pkl', 'wb'))
```
