```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('credit_train.csv')
```

```python
df.head(5)
```

```python
df['Loan Status'].value_counts()
```

```python
df2 = df.drop(['Loan ID','Customer ID'],axis=1)
```

```python
df['Term'].value_counts()
```

```python
df['Purpose'].value_counts()
```

```python
df2.head(5)
```

```python
df2 = pd.concat([df2.drop('Years in current job', axis =1 ), pd.get_dummies(df2['Years in current job'], prefix='Years in current job')], axis=1)
df2 = pd.concat([df2.drop('Home Ownership', axis =1 ), pd.get_dummies(df2['Home Ownership'], prefix='Home Ownership')], axis=1)
df2 = pd.concat([df2.drop('Purpose', axis =1 ), pd.get_dummies(df2['Purpose'], prefix='Purpose')], axis=1)
df2['Term'] = df2['Term'].apply(lambda x: 1 if x == 'Short Term' else 0)
df2['Loan Status'] = df2['Loan Status'].apply(lambda x: 1 if x == 'Fully Paid' else 0)

```

```python
df2.head(5)
```

```python
df2.fillna(0, inplace=True)
```

```python
df2.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

train, test = train_test_split(df2, test_size=0.2, random_state=42)
```

```python
train_x = train.drop('Loan Status', axis =1)
train_y = train['Loan Status']
test_x = test.drop('Loan Status', axis =1)
test_y = test['Loan Status']
```

```python
train_x.isna().sum().sum()
```

```python
forest = RandomForestClassifier(n_estimators=100)
forest.fit(train_x, train_y)
```

```python
forest.score(test_x, test_y)
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [10, 20, 30, None],
    'max_features': ['auto', 'sqrt', 'log2'],
    'min_samples_split': [2, 5, 10],
}

# Using RandomizedSearchCV
random_search = RandomizedSearchCV(forest, param_distributions=param_grid,
                                   n_iter=20, cv=3, verbose=2, scoring='accuracy', random_state=42)

# Fitting the model
random_search.fit(train_x, train_y)
```

```python
random_search.best_estimator_
```

```python
random_search.best_score_
```

```python
import pickle
pickle.dump(forest, open('loansmodel.pkl', 'wb'))
```

```python
test_unknown = pd.read_csv('credit_test.csv')
```

```python
test_unknown.head(5)
```

```python
test_unknown2 = test_unknown.drop(['Loan ID','Customer ID'],axis=1)
```

```python
test_unknown2 = pd.concat([test_unknown2.drop('Years in current job', axis =1 ), pd.get_dummies(test_unknown2['Years in current job'], prefix='Years in current job')], axis=1)
test_unknown2 = pd.concat([test_unknown2.drop('Home Ownership', axis =1 ), pd.get_dummies(test_unknown2['Home Ownership'], prefix='Home Ownership')], axis=1)
test_unknown2 = pd.concat([test_unknown2.drop('Purpose', axis =1 ), pd.get_dummies(test_unknown2['Purpose'], prefix='Purpose')], axis=1)
test_unknown2['Term'] = test_unknown2['Term'].apply(lambda x: 1 if x == 'Short Term' else 0)
```

```python
# Load the trained model
loaded_model = pickle.load(open('loansmodel.pkl', 'rb'))

test_unknown2.fillna(0, inplace=True)

# Make predictions on the test_unknown dataset
predictions = loaded_model.predict(test_unknown2)
```

```python
predictions
```

```python
# Create a DataFrame from the predictions array
predictions_df = pd.DataFrame(predictions, columns=['Status'])
```

```python
# Replace 1 with "Fully Paid" and 0 with "Charged Off"
predictions_df['Status'] = predictions_df['Status'].replace({1: 'Fully Paid', 0: 'Charged Off'})
```

```python
merge_df = pd.concat([test_unknown['Loan ID'], predictions_df], axis=1)
```

```python
merge_df.head(5)
```

```python
# Save the merged DataFrame to a CSV file
csv_file_path = 'loan_predictions.csv'
merge_df.to_csv(csv_file_path, index=False)
```
